```python
def intersection(nums1, nums2):
    # Create empty dictionary and result list
    d = {}
    result = []
    
    # Count occurrences of each number in nums1
    for num in nums1:
        if num in d:
            d[num] += 1
        else:
            d[num] = 1
    
    # Check for common numbers in nums2
    for num in nums2:
        if num in d and d[num] > 0:
            result.append(num)
            d[num] -= 1
    
    # Return result list
    return result

# Get input arrays from user
nums1 = list(map(int, input().split()))
nums2 = list(map(int, input().split()))

# Compute intersection and print result
result = intersection(nums1, nums2)
for num in result:
    print(num, end=' ')
```

```python
def is_matching(s, p):
    if len(s) == 0 and len(p) == 0:  # edge case where both strings are empty
        return True
    elif len(p) == 0:  # edge case where p is empty
        return False
    elif len(s) == 0:  # edge case where s is empty but p contains only '*' characters
        return all(c == '*' for c in p)
    elif p[0] == '?':  # if the first character of p is '?', it matches the first character of s
        return is_matching(s[1:], p[1:])
    elif p[0] == '*':
        return any(is_matching(s[i:], p[1:]) for i in range(len(s)+1))  # check if the rest of s matches the rest of p after the '*'
    else:
        return s[0] == p[0] and is_matching(s[1:], p[1:])

# example usage
s = 'aa'
p = 'a'
print(is_matching(s, p))  # False

s = 'aa'
p = 'a?'
print(is_matching(s, p))  # True

s = 'aa'
p = '*'
print(is_matching(s, p))  # True

s 
```

```python
class Solution:
    def doesitatch(self, s, p, memo=None):
        if memo is None:
            memo = {}
        if (s, p) in memo:
            return memo[(s, p)]

        if len(s) == 0 and len(p) == 0:
            memo[(s, p)] = True
        elif len(p) == 0:
            memo[(s, p)] = False
        elif len(s) == 0 and all(c == '*' for c in p):
            memo[(s, p)] = True
        elif p[0] == '?':
            memo[(s, p)] = self.doesitatch(s[1:], p[1:], memo)
        elif p[0] == '*':
            memo[(s, p)] = any(self.doesitatch(s[i:], p[1:], memo) for i in range(len(s)+1))
        else:
            memo[(s, p)] = s[0] == p[0] and self.doesitatch(s[1:], p[1:], memo)
        return memo[(s, p)]

def main():
    s = input()
    p = input()
    print(Solution().doesitatch(s, p))

if __name__ == '__main__':
    main()
```

```python
import math
import os
import random
import re
import sys

# Complete the 'containsAllCharacters' function below.
# The function is expected to return a STRING_ARRAY.
# The function accepts STRING_ARRAY given_strings as parameter.

def containsAllCharacters(given_strings):
    # Write your code here
    
    # iterate through each string in the given_strings array
    results = []
    for s in given_strings:
        # create sets for upper case and lower case English letters
        upper = set('ABCDEFGHIJKLMNOPQRSTUVWXYZ')
        lower = set('abcdefghijklmnopqrstuvwxyz')
    
        # iterate through the string and add each character to the corresponding set
        for char in s:
            if char.isupper():
                upper.discard(char)
            elif char.islower():
                lower.discard(char)
    
        # check if both sets have size 0, which means they contain all the English alphabet characters
        if len(upper) == 0 and len(lower) == 0:
            results.append("YES")
        else:
            results.append("NO")
    
    return results

if __name__ == '__main__':
    given_strings_count = int(input().strip())
    given_strings = []
    for _ in range(given_strings_count):
        given_strings_item = input()
        given_strings.append(given_strings_item)

    results = containsAllCharacters(given_strings)

    for result in results:
        print(result)
```
